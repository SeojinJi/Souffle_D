```python
# factorial function

n = int(input())

def factorial(n):
    if n == 0 :
        return 1
    else :
        return n * factorial(n-1)
    
print(factorial(n))
```

```python
# Fibonacci

n = int(input())

fibo = [0] * (n+1)

for i in range(n+1):
    if i == 0 :
        fibo[i] = 0
    elif i == 1 :
        fibo[i] = 1
    else :
        fibo[i] = fibo[i-1] + fibo[i-2]
        
print(fibo[n])
```

```python
# making star (time)
n = int(input())

def star(n):
    if n == 3 :
        stargazer = [['*', '*', '*'], ['*', '  ', '*'], ['*', '*', '*']]
        return stargazer
    else :
        stargazer = []
        num = n//3
        for j in range(3) :
            for i in range(num):
                stargazer.append(star(num)[i] * 3)
                
        for i in range(num, 2*num):
            for j in range(num, 2*num):
                stargazer[i][j] = '  '
                
        return stargazer
    
stargazer = star(n)

for i in range(len(stargazer)):
    for j in range(len(stargazer[i])):
        print(stargazer[i][j], end = '')
    print()
```
